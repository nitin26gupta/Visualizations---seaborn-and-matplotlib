## 2b Context Functions

```python
# Paper
# Notebook
# Talk
# Poster
# By default, context is set to notebook;

f = plt.figure(figsize=(10, 10))
gs = f.add_gridspec(2, 2)


sns.set_context("poster", # Set context to "poser"
                    font_scale=1.5, # Scale font elements 
                    rc={"lines.linewidth":2.5}) # and override param mapping
ax = f.add_subplot(gs[0, 0])
sinplot()

sns.set_context("talk", # Set context to "talk"
                    font_scale=1.5, # Scale font elements 
                    rc={"lines.linewidth":2.5}) # and override param mapping
ax = f.add_subplot(gs[0, 1])
sinplot()

sns.set_context("notebook", # Set context to "notebook"
                    font_scale=1.5, # Scale font elements 
                    rc={"lines.linewidth":2.5}) # and override param mapping
ax = f.add_subplot(gs[1, 0])
sinplot()

sns.set_context("paper", # Set context to "paper"
                    font_scale=1.5, # Scale font elements 
                    rc={"lines.linewidth":2.5}) # and override param mapping
ax = f.add_subplot(gs[1, 1])
sinplot()

f.tight_layout()
```
